# EDA

## Univariate Analysis

**Ship Mode**

```python
ship_mode = supermarket['Ship Mode'].value_counts()
print(ship_mode)
```

```python
# Number of Products by Ship Mode

plt.figure(figsize=(7,5))
ship_mode.plot(kind='bar')
plt.title('Number of Product by Ship Mode')
plt.xlabel('Ship Mode')
plt.ylabel('Number of Product')
plt.show()
```

**Segment**

```python
segment = supermarket['Segment'].value_counts()
print(segment)
```

```python
# % of Product by Segment

plt.figure(figsize=(7,5))
segment.plot(kind='pie', autopct='%1.1f%%')
plt.title('Product Percentage by Segment')
plt.ylabel('')
plt.show()
```

**Region**

```python
region = supermarket['Region'].value_counts()
print(region)
```

```python
# % of product by Region

plt.figure(figsize=(7,5))
region.plot(kind='pie', autopct='%1.1f%%')
plt.title('Product Percentage by Region')
plt.ylabel('')
plt.show()
```

**Category**

```python
category = supermarket['Category'].value_counts()
print(category)
```

```python
# Number of product by Category

plt.figure(figsize=(7,5))
category.plot(kind='bar', color='maroon')
plt.title('Number of Product by Category')
plt.xlabel('Category')
plt.ylabel('Number of Product')
plt.xticks(rotation=0)
plt.show()
```

**Sub-Category**

```python
sub_category = supermarket['Sub-Category'].value_counts()
print(sub_category)
```

```python
# Number of product by Sub-Category

plt.figure(figsize=(10,5))
sub_category.plot(kind='bar', color='navy')
plt.title('Number of Product by Sub-Category')
plt.xlabel('Sub-Category')
plt.ylabel('Number of Product')
plt.show()
```

## Bivariate Analysis

**1. Sales vs Ship Mode**

```python
shipMode_grp = supermarket.groupby('Ship Mode')['Sales'].sum()
print(shipMode_grp)
```

```python
# Sales by Ship Mode in %

shipMode_grp.plot(kind='pie', autopct='%1.1f%%')
plt.title('Sales by Ship Mode')
plt.ylabel('')
plt.show()
```

**2. Sales vs Segments**

```python
segment_grp = supermarket.groupby('Segment')['Sales'].sum()
print(segment_grp)
```

```python
# Sales by Segment in %

segment_grp.plot(kind='pie', autopct='%1.1f%%')
plt.title('Sales by Segment')
plt.ylabel('')
plt.show()
```

**3. Top 10 Cities by Sales**

```python
# Top 10 Cities by total Sales

city_grp = supermarket.groupby('City')['Sales'].sum().sort_values(ascending=False).head(10)
print(city_grp)
```

```python
# To 10 cities by total sales

city_grp.plot(kind='barh', color='navy')
plt.title('Top 10 Cities by Sales')
plt.xlabel('Total Sales')
plt.ylabel('City')
plt.show()
```

**4. Top 10 states by total sales**

```python
state_grp = supermarket.groupby('State')['Sales'].sum().sort_values().head(10)
print(state_grp)
```

```python
# Top 10 States by total Sales

state_grp.plot(kind='barh', color='purple')
plt.title('Top 10 States by Sales')
plt.xlabel('Total Sales')
plt.ylabel('State')
plt.show()
```

**5. Sales Percentage by Region**

```python
region_grp = supermarket.groupby('Region')['Sales'].sum()
print(region_grp)
```

```python
# Sales by Region in %

region_grp.plot(kind='pie', autopct='%1.1f%%')
plt.title('Seles percentage by Region')
plt.ylabel('')
plt.show()
```

**6. Sales percentage by Category**

```python
category_grp = supermarket.groupby('Category')['Sales'].sum()
print(category_grp)
```

```python
# Sales by Category in %

category_grp.plot(kind='pie', autopct='%1.1f%%')
plt.title('Sales percentage by Category')
plt.ylabel('')
plt.show()
```

**7. Total Sales by Sub-Catgegory**

```python
sub_category_grp = supermarket.groupby('Sub-Category')['Sales'].sum().sort_values()
print(sub_category_grp)
```

```python
sub_category_grp.plot(kind='bar', color='green')
plt.title('Total Sales by Sub-Category')
plt.xlabel('Sub-Category')
plt.ylabel('Total Sales')
plt.show()
```
